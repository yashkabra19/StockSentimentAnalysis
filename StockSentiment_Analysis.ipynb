```python
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.ensemble import RandomForestClassifier
```

```python
vectorizer = TfidfVectorizer(ngram_range = (2,2))
traindataset = vectorizer.fit_transform(headlines)
```

```python
randomclassifier=RandomForestClassifier(n_estimators=200,criterion='entropy')
randomclassifier.fit(traindataset,train['Label'])
```

```python
test_data = test.iloc[: , 2 :27]
test_data.replace("[^a-zA-Z0-9]", " " , regex = True, inplace = True)
list2 = [(i) for i in range(25)]
New_index = [str(i) for i in list2]

test_data.columns = New_index
test_data.head()

for index in New_index :
    test_data[index] = test_data[index].str.lower()
    
```

```python
 test_transform= []
for row in range(0,len(test.index)):
    test_transform.append(' '.join(str(x) for x in test.iloc[row,2:27]))
test_dataset = vectorizer.transform(test_transform)
predictions = randomclassifier.predict(test_dataset)
```

```python
 from sklearn.metrics import classification_report,confusion_matrix,accuracy_score
```

```python
matrix=confusion_matrix(test['Label'],predictions)
print(matrix)
score=accuracy_score(test['Label'],predictions)
print(score)
report=classification_report(test['Label'],predictions)
print(report)
```
